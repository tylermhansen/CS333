```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv("merge_timing.csv", delimiter="\t")
df.columns = df.columns.str.replace(' ', '')
df.plot('Threads', 'Time', figsize=(16,8), title="Merge Sort Time vs. Number of Threads", ylabel="Merge Sort time (sec)")
plt.savefig("MergeSort.png")
```
