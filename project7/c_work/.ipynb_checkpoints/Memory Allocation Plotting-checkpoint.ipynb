```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv("times.csv", delimiter="\t")
df.columns = df.columns.str.replace(' ', '')
df.plot('Bytes', 'malloc_time', figsize=(16,8), title="malloc() time vs. Number of Bytes to Allocate", ylabel="malloc() time")
plt.show()
```

```python
df.plot('Bytes', 'free_time', figsize=(16,8), title="free() time vs. Number of Bytes to Free from Memory", ylabel="free() time")
plt.show()
```
